```python
import pandas as pd
data=pd.read_csv('loan_data.csv')
```

```python
data
```

```python
field_names=data.columns
field_names
```

```python
label=['male','female']
male=data[data[field_names[1]]=='Male']['Gender'].count()
female=data[data[field_names[1]]=='Female']['Gender'].count()
freq=[male,female]
plt.pie(freq,labels=label,autopct="%1.1f%%")
```

```python
label=['Married','Unmarried']
married=data[data[field_names[2]]=='Yes']['Married'].count()
unmarried=data[data[field_names[2]]=='No']['Married'].count()
freq=[married,unmarried]
plt.pie(freq,labels=label,autopct="%1.1f%%")
```

```python
dependents_unique=data[field_names[3]].drop_duplicates().to_list()
freq=[]
for i in dependents_unique:
    freq.append(  data[data[field_names[3]] == i]['Dependents'].count()  )
dependents_unique[4]='NAN'
dependents_unique.sort()
plt.bar(dependents_unique, height = np.array(freq))
plt.ylabel('frequency')
plt.xlabel('Dependents')
```

```python
dependents=data[[field_names[3],field_names[8]]]
mask = dependents[(dependents[field_names[3]] != '3+')].dropna()
plt.scatter(mask.iloc[:,0], mask.iloc[:,1])
print(mask)
```

```python
plt.boxplot([float(var) for var in mask.iloc[:,1]])
plt.yticks(range(0,160,10))
plt.ylabel('Loan Amount')
```

```python
graduates=data[ data[field_names[4]] == "Graduate" ]['Education'].count()
ungraduates=data[data[field_names[4]]=="Not Graduate"]['Education'].count()
plt.bar(['graduate','non graduate'],[graduates,ungraduates])
```

```python
mask=data[[field_names[6],field_names[8]]]
plt.scatter(mask.iloc[:,0], mask.iloc[:,1])
plt.ylabel('Loan Amount')
plt.xlabel('Income')
```
