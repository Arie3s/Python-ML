```python
loan_data=pd.read_csv('loan_data.csv')
```

```python
loan_data.dropna(inplace=True)
```

```python
loan_data
```

```python
approved=loan_data[loan_data['Loan_Status']=='Y']
self_employed=approved[approved['Self_Employed']=='Yes']['Self_Employed'].count()
employed=approved[approved['Self_Employed']=='No']['Self_Employed'].count()
plt.pie([self_employed,employed],labels=['self employed','employed'])
```

```python
not_approved=loan_data[loan_data['Loan_Status']=='N']
self_employed=not_approved[not_approved['Self_Employed']=='Yes']['Self_Employed'].count()
employed=not_approved[not_approved['Self_Employed']=='No']['Self_Employed'].count()
plt.pie([self_employed,employed],labels=['self employed','employed'])
```
