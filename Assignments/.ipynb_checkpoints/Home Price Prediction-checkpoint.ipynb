```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from numpy import linalg as lalg
import seaborn as sns
```

```python
data = pd.read_csv('house_data.csv')
```

```python
data
```

```python
data.hist(figsize=(15,10))
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
data = data.drop(columns=['page_url','latitude','longitude'],axis=0)
```

```python
data = data.drop(columns=['agent','agency','Unnamed: 0'],axis=0)
```

```python
data = data.drop(columns=['date_added'],axis=0)
```

```python
data
```

```python
data.drop(columns=['property_type'],axis=0).duplicated().unique()
```

```python
data
```

```python
grouped = data[data['property_type'] == 'House'].groupby(['city','location'])
```

```python
citywise = {}
for (city, location), group in grouped:
    if city in citywise:
        citywise[city][location] = group
    else:
        citywise[city] = {location: group}
```

```python
print(citywise.keys())
```

```python
# Assuming citywise is your dictionary
citywise_updated = {}
for city, location_data in citywise.items():
    city_data_updated = {}
    for location, data in location_data.items():
        # Convert data to a DataFrame and remove 'city' and 'location' columns
        data_df = pd.DataFrame(data)
        city_data_updated[location] = data_df
    citywise_updated[city] = city_data_updated
citywise = citywise_updated
```

```python
bahria_town_karachi = citywise['Karachi']['Bahria Town Karachi']
```

```python
btown_khi_house_sale = bahria_town_karachi[bahria_town_karachi['purpose'] == 'For Sale'].drop(columns = ['purpose','property_type'])
```

```python
btown_khi_house_sale
```

```python
plt.scatter(btown_khi_house_sale['Total_Area'], btown_khi_house_sale['price'])
plt.xlabel('Total Area')
plt.ylabel('Price')
plt.title('Scatter Plot of Total Area vs Price')
plt.show()
```

```python
plt.scatter(btown_khi_house_sale['Total_Area'], btown_khi_house_sale['price'])
plt.xlabel('Total Area')
plt.ylabel('Price')
plt.title('Scatter Plot of Total Area vs Price')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = btown_khi_house_sale['Total_Area'].to_numpy().reshape(-1,1)
y = btown_khi_house_sale['price'].to_numpy()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
print(x_train.shape,y_train.shape)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R^2 Score:", r2)
```

```python
for i, j, k in zip(x_test.flatten(), y_pred.flatten(), y_test.flatten()):
    print(f'Area : {i:.2f} prediction : {j:.2f}  actual : {k:.2f}\n')
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2)
x_train_quadratic = poly_features.fit_transform(x_train)
x_test_quadratic = poly_features.transform(x_test)
```

```python
model_quadratic = LinearRegression()
model_quadratic.fit(x_train_quadratic, y_train)
```

```python
y_pred_quadratic = model_quadratic.predict(x_test_quadratic)
```

```python
for i, j, k in zip(x_test.flatten(), y_pred_quadratic.flatten(), y_test.flatten()):
    print(f'Area : {i:.2f} prediction : {j:.2f}  actual : {k:.2f}\n')
```

```python
model_quadratic.score(x_test_quadratic,y_test)
```
