# Numpy - Numerical Python
- NumPy is a fundamental package for scientific computing in Python. 
- It provides support for multi-dimensional arrays and matrices, along with a variety of mathematical functions to operate on these arrays. 
- Performance: NumPy is implemented in C and Fortran, making it fast and efficient, especially for large datasets.
- Open Source: NumPy is an open-source project and is actively developed and maintained by a community of contributors.
- Integration with Other Libraries: NumPy is a foundational library for many other scientific computing libraries in Python, such as SciPy, Pandas, Matplotlib, and scikit-learn.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

- Arrays: NumPy arrays are similar to Python lists but are more efficient for numerical computations.
- Data Types: NumPy arrays have a data type (dtype) associated with them, such as int, float, complex, etc.


```python
arr1=np.array([2,5,6,9], dtype='int8')
arr1.shape, arr1.dtype
```

```python
arr2=np.array([[3.5, 4.0, 6.5],[0.4, 0.9, 4.7]])
arr2.shape
```

- Vectorized Operations: NumPy supports vectorized operations, allowing mathematical operations to be performed on entire arrays without the need for explicit looping in Python.



```python
a1=np.array([1,2,3,4])
a2=np.array([10,20,30,40])
a1*a2
```

```python
a1*100
```

- Broadcasting: NumPy arrays support broadcasting, which allows arrays of different shapes to be combined in arithmetic operations.


- Indexing and Slicing: NumPy provides powerful indexing and slicing capabilities for accessing elements or subarrays of arrays.


- Universal Functions (ufuncs): NumPy includes a large collection of mathematical functions called universal functions (ufuncs) that operate element-wise on arrays.


```python
import time
```

```python
a1=np.arange(1,100000000)
a2=np.arange(1,100000000)
st=time.time()
a3=a1*a2
end=time.time()-st
print("time in sec using * operator: ", end)
st=time.time()
np.multiply(a1,a2)
end=time.time()-st
print("time in sec using np.multiply: ", end)
```

- Linear Algebra: NumPy provides functions for linear algebra operations such as matrix multiplication, inversion, eigenvalue decomposition, and solving linear systems.
 

```python
np.dot([2j, 3j], [2j, 3j])
```

```python
a = [[1, 0], [0, 1]]
b = [[4, 1], [2, 2]]
np.dot(a, b)
```

# Plots for pairwise data
- Pairwise data refers to a dataset where each observation consists of pairs of values. These pairs could represent various types of relationships or measurements between two entities. 
- Pairwise data is commonly used in many fields such as statistics, machine learning, and social sciences. 
## 1. Line Plot: 
- Used to visualize data points connected by straight line segments.

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 3, 5, 7, 11, 13, 15, 17, 19]

x1 = np.linspace(0, 10, 30)
y1 = 4 + 2 * np.sin(2 * x1)

plt.figure(figsize=(4,4))
plt.plot(x, y, '-.o', label='simple line')
plt.plot(x1,y1, ':s', label='sine wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
```

## Scatter Plot: 
- Used to visualize the relationship between two variables by displaying individual data points.

```python
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

plt.figure(figsize=(6,6))
plt.scatter(x, y, c=colors)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.colorbar()
```

## Bar Plot: 
- Used to compare categorical data by displaying bars of different heights.

```python

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.figure(figsize=(4,4))
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
```

## fill_between plot:
- It allows you to shade the area between two curves or between a curve and a reference line. 

```python
# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot the curves
plt.plot(x, y1, color='blue', label='sin(x)')
plt.plot(x, y2, color='red', label='cos(x)')

# Shade the area between the curves
plt.fill_between(x, y1, y2, color='gray', alpha=0.3)

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fill Between Plot')
plt.legend()

```

## Statistical Plots
# Histogram: 
- Used to visualize the distribution of a continuous variable.
- useful for visualizing the distribution of a dataset by displaying the frequencies or counts of data points within different bins or intervals. 

```python
data = np.random.randn(1000)
plt.figure(figsize=(5,5))
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3
plt.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)));
plt.colorbar()
```

## Pie Chart: 
- Used to display the proportions of different categories as slices of a pie.
- They are useful for visually comparing the relative sizes of different categories or showing how a whole is divided into parts. 

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
```

## Boxplot: 
- Also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset along with key statistical measures. 
- It provides a concise summary of the data's central tendency, spread, and presence of outliers.

```python
# Generate some random data
np.random.seed(10)
data = np.random.normal(loc=0, scale=1, size=100)

# Create a boxplot
plt.boxplot(data)

# Add labels and title
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Boxplot Example')
```

## Errorbar plot:
- It is used to visualize the uncertainty or variability in data points by displaying error bars around each data point. 
- These error bars represent the range of uncertainty or variability associated with each data point. 

```python
# Generate some random data
x = np.arange(10)
y = np.sin(x)
errors = np.random.uniform(0.1, 0.3, size=10)  # Random error values

# Create an errorbar plot
plt.errorbar(x, y, yerr=errors, fmt='o', capsize=5)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Errorbar Plot')
```
