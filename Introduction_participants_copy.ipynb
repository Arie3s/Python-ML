## List
  - Implemented as an array of pointers to objects therefore the list can store elements of different types
  - The underlying array is dynamically resized as elements are added or removed from the list
  - This array-based implementation makes certain operations, such as indexing and appending, efficient.
  - However, it's important to note that the array can become fragmented over time due to the resizing operations, which may impact the overall performance in terms of memory usage.
  - Mutable

## Dictionary
  - A python data structure that stores values in key:value format (like word:meaning)
  - Dictionaries are implemented in python as hash tables.
  - A hash table is a data structure that allows for efficient insertion, deletion, and retrieval of key-value pairs.
  - The dictionary implementation in Python uses a combination of hashing and open addressing to achieve these operations with good average time complexity.
  - Common open addressing methods include linear probing (checking the next position), quadratic probing (checking positions with quadratic increments), and double hashing (using a secondary hash function to calculate the next index).
  - In addition to open addressing, Python also uses a technique known as separate chaining to handle collisions. Each slot in the hash table can store not just one key-value pair, but a linked list (or other data structure) of key-value pairs that have the same hash value.
  - Values in a dictionary can be of any data type and can be duplicated, whereas keys can't be repeated and must be immutable.

## Tuple
  - In Python, tuples are implemented as fixed-size arrays that contain references to objects.
  - Tuples are similar to lists, but unlike lists, they are immutable, meaning their elements cannot be changed or modified after the tuple is created.
  - The implementation details ensure that tuples are more memory-efficient compared to lists for certain use cases.
  - Tuples can contain elements of different data types, and each element is a reference to an object.
  - Tuples are generally faster than lists for certain operations because of their immutability. For example, iterating over a tuple can be faster than iterating over a list, as the interpreter doesn't need to check for changes during iteration.
  - Tuple creation is generally faster than list creation because there is no need to allocate extra memory for potential future modifications.
  

```python
tup = (1, 2, 'hello', 3.14, 1)
tup
```

```python
tup.index(2)
```
